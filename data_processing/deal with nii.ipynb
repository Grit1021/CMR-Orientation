```python
# 主函数
def nii_to_image(filepath):
    # 读取nii文件夹
    filenames = os.listdir(filepath)
    
    # 开始读取nii文件
    for f in filenames:
        img_path = os.path.join(filepath, f)
        img = nib.load(img_path)  # 读取nii
        img_fdata = img.get_fdata()
        # 去掉nii的后缀名
        fname = f.replace('.nii', '')  
        img_f_path = os.path.join(imgfile, fname)

        # 创建nii对应的图像的文件夹
        if not os.path.exists(img_f_path):
            # 新建文件夹
            os.mkdir(img_f_path)  

        # 开始转换为图像
        (x, y, z) = img.shape
        # z是图像的序列
        for i in range(z):
            # 选择哪个方向的切片都可以
            silce = img_fdata[:, :, i]
            # 保存图像
            imageio.imwrite(os.path.join(img_f_path, '{}.png'.format(i)), silce)

if __name__ == '__main__':
    filepath = 'D:/2021-2022/SDS/postgrad/ZMIC/Orientation-Adjust-Tool-master/data/valid_data_LGE'
    imgfile = 'D:/2021-2022/SDS/postgrad/ZMIC/Orientation-Adjust-Tool-master/data/valid_data_LGE'
    nii_to_image(filepath)
```
